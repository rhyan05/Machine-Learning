# Exemplo simples para classificação


Caracteristica de animais
Porco e cachorro

1. Pelo longo
2. Perna Curta
3. Faz au au





imagem_2025-02-19_084040217.png

```python
# Porco
porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]

# Cachorro
cachorro1 = [0,1,1]
cachorro2 = [1,0,1]
cachorro3 = [1,1,1]

# Dados para treinamento

dados =[porco1, porco2, porco3, cachorro1,cachorro2,cachorro3]

# Classifificação

classses = [1,2,1,0,0,0]
```

```python
# importa pacote de algortimo para ML
# sklearn.sum(biblioteca - recursos)
#classe- LinearSVC: classificaçao

from sklearn.svm import LinearSVC
classes = [1,2,1,0,0,0]

model = LinearSVC()

# Analisar os dados usando a função mode.fit(...)
model.fit(dados, classes)
```

```python
# Realizando os testes
misterio1 = [1,1,1]
misterio2 = [0,1,0]
misterio3 = [0,1,1]


testes = [misterio1, misterio2, misterio3]
previsoes = model.predict(testes)
print(previsoes)
```

```python
testes_classes = [0,1,1]

#fazendo comparaçao
comparacao =(previsoes == testes_classes)
print(comparacao)
```

```python
# Somar os corretos

corretos = (previsoes == testes_classes).sum()
print(type(corretos))

print(corretos)
```

```python
# Tamanha do teste
total = len(testes)

# Taxa de acertos
taxa_acerto = corretos / total
print(f"taxa de acerto :{taxa_acerto*100:.2f}")
```

```python
# Utilizando os recursos da biblioteca para calcular a porcentagem de acerto
from sklearn.metrics import accuracy_score
taxa_acerto = accuracy_score(testes_classes, previsoes)
print(f"taxa de acerto :{taxa_acerto*100:.2f}")
```
